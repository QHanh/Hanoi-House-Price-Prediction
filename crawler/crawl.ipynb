```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
LIST_LINK_PRODUCT = []

for i in range(2000):
    try:
        response = requests.get("https://bds.com.vn/mua-ban-nha-dat-ha-noi-page" + str(i + 1))
        soup = BeautifulSoup(response.content, "html.parser")
        list_link = soup.findAll('div', 'item-nhadat')          
        for item in list_link:
            link = item.find('a', 'title-item-nhadat').get('href')
            LIST_LINK_PRODUCT.append(
                {
                    'link': link
                }
            )
        print('Done page ' + str(i + 1))
    except Exception as e:
        print('Error get link ' + str(i + 1))
        print(e)
```

```python
LIST_PRODUCT = []

for i, item in enumerate(LIST_LINK_PRODUCT):
    link = item['link']
    try:
        response = requests.get(link)
        soup = BeautifulSoup(response.content, "html.parser")

        try:
            title = soup.find('h1', 'title-product').text.strip()
        except Exception as e:
            print(f'Error extracting title: {e}')
            title = None

        try:
            infoAddress = soup.find('ul', 'breadcrumb').findAll('li')
            estate_type = ' '.join(infoAddress[0].text.split()[2:]) if len(infoAddress) > 0 else None
            province = infoAddress[1].text if len(infoAddress) > 1 else None
            district = infoAddress[2].text if len(infoAddress) > 2 else None
            ward = infoAddress[3].text if len(infoAddress) > 3 else None
        except Exception as e:
            print(f'Error extracting infoAddress: {e}')
            estate_type = province = district = ward = None

        try:
            main_info = soup.findAll('ul', 'list-attr-hot')[0].findAll('li')
            price = main_info[0].findAll('span')[1].text.strip() if len(main_info) > 0 else None
            square = main_info[1].findAll('span')[1].text.strip() if len(main_info) > 1 else None
            post_date = main_info[2].findAll('span')[1].text.strip() if len(main_info) > 2 else None
        except Exception as e:
            print(f'Error extracting important_info: {e}')
            price = square = post_date = None

        try:
            describe = soup.find('div', 'ct-pr-sum').text.strip()
        except Exception as e:
            print(f'Error extracting describe: {e}')
            describe = None

        numb_bedroom = numb_toilet = numb_floor = None
        try:
            if len(soup.findAll('ul', 'list-attr-hot')) > 1:
                extent_info = soup.findAll('ul', 'list-attr-hot')[1].findAll('li')
                for ele in extent_info:
                    label = ele.findAll('span')[0].text
                    value = ele.findAll('span')[1].text.strip()
                    if label == 'S phng ng :':
                        numb_bedroom = value
                    elif label == 'S toilet :':
                        numb_toilet = value
                    elif label == 'Tng :':
                        numb_floor = value
        except Exception as e:
            print(f'Error extracting extent_info: {e}')

        LIST_PRODUCT.append({
            'Link': link,
            'Title': title,
            'Estate_type': estate_type,
            'Province': province,
            'District': district,
            'Ward': ward,
            'Price': price,
            'Square': square,
            'Post_date': post_date,
            'Describe': describe,
            'Numb_bedroom': numb_bedroom,
            'Numb_toilet': numb_toilet,
            'Numb_floor': numb_floor,
        })

        print(f'Done {i + 1}')
    except Exception as e:
        print(f'Error {i + 1}')
        print(e)
```

```python
df_link_product = pd.DataFrame(LIST_PRODUCT)
df_link_product.to_csv('data/list_house.csv')
```
